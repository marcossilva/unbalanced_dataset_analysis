```python
def oversample(df, times=20):
    df_new_ratio = df
    for i in range(times):
        df_new_ratio = df_new_ratio.append(df[df.failure == 1])
        ratio = df_new_ratio.failure.value_counts()
    print(f'ratio[0]/ratio[1] = {ratio[0]}/{ratio[1]} = {ratio[0]/ratio[1]}')
    return df_new_ratio
```

```python
from sklearn.model_selection import train_test_split
def split(df):
    X = df.loc[:, ["attribute1","attribute2","attribute3","attribute4","attribute5","attribute6","attribute7","attribute8","attribute9"]]
    Y = df.loc[:, ["failure"]]
    return train_test_split(X,
                            Y.values.reshape(-1, 1).ravel(),
                            test_size=0.33)
```

```python
df_new_ratio = oversample(df, times=1)
X_train, X_test, y_train, y_test = split(df_new_ratio)
```

## Raw Models

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
def print_report(clf, X_test=X_test, y_test=y_test):
    y_pred = clf.predict(X_test)
    print(classification_report(y_test, y_pred))
    #     Note that in binary classification, recall of the positive class
    #     is also known as "sensitivity"; recall of the negative class is
    #     "specificity".
    print(f'Acc: {accuracy_score(y_test,y_pred)*100}')
    print(f'AUC: {roc_auc_score(y_test,y_pred)}')
    print(confusion_matrix(y_test, y_pred))
    fpr, tpr, _ = roc_curve(y_test, y_pred)
    plt.figure(1)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot(fpr, tpr, label='Classifier')
    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.title('ROC curve')
    plt.legend(loc='best')
    plt.show()
    
    plt.figure(2)
    plt.xlim(0, 0.1)
    plt.ylim(0.9, 1)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot(fpr, tpr, label='Classifier')
    plt.xlabel('False positive rate')
    plt.ylabel('True positive rate')
    plt.title('ROC curve (zoomed in at top left)')
    plt.legend(loc='best')
    plt.show()
```
